```python
#cleaming data
backup = data.copy()
```

```python
data=data[data['year'].str.isnumeric()]
```

```python
data['year']=data['year'].astype(int)
```

```python
data=data[data['Price']!='Ask For Price']
```

```python
data['Price']=data['Price'].str.replace(',','').astype(int)
```

```python
data['kms_driven']=data['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
data[data['kms_driven']!='Petrol']
```

```python
data=data[data['kms_driven'].str.isnumeric()]
```

```python
data['kms_driven']=data['kms_driven'].astype(int)
```

```python
data
```

```python
data=data[~data['fuel_type'].isna()]
```

```python
data['name']=data['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
data=data.reset_index(drop=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
data=data[data['Price']<6e6].reset_index(drop=True)
```

```python
data
```

```python
data.to_csv('cleaned_car_data.csv')
```

```python
#model
```

```python
X=data.drop(columns=['Price'])
y=data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','fuel_type','company']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','fuel_type','company']),remainder='passthrough')
```

```python
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
score=[]
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    score.append(r2_score(y_test,y_pred))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=np.argmax(score))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
print(r2_score(y_test,y_pred))
```

```python
import pickle
pickle.dump(pipe,open('car_pricing_linear_reg.pkl','wb'))
```
